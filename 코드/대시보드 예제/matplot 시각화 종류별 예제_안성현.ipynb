```python
import matplotlib.pyplot as plt
import pandas as pd
import os
currentPath=os.getcwd()
print(currentPath)

file=currentPath+"\Report2.csv"
traffic=pd.read_csv(file , encoding = "utf-8")
```

```python
traffic = traffic.loc[:,['기간','지역','발생건수','사망자수','부상자수']]
print(traffic)
```

```python
drop_list = traffic[traffic['지역'].str.contains('합계')].index #지역에 합계인 행을 찾아 인덱스 반환
traffic.drop(drop_list, inplace=True) # 행 버리기
print(traffic)
```

```python
# 발생건수에 , 없애기
traffic['발생건수'] = traffic['발생건수'].str.replace('[,]','') 
# 발생건수 숫자로 변환
traffic = traffic.astype({'발생건수':'int' }) 
traffic['부상자수'] = traffic['부상자수'].str.replace('[,]','')
traffic = traffic.astype({'부상자수':'int' })
```

```python
traffic.info()
#traffic.reset_index(inplace=True)
print(traffic)
```

```python
found= traffic['지역'] == '중구'
junggu =  traffic[found]
print(junggu)
```

#### 산점도 그리기

#### 히스토그램 그리기
x축은 종류, y축은 빈도, 개수 등
plt.hist(X축 틱 배열)  
bins : 가로축 구간의 개수 지정  
histtype : 히스토그램의 종류. bar, barstacked, step, stepfilled  
density : True 로 설정하면 밀도 함수가 되어 막대의 아래 면적이 1이 됨

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.hist((weight, weight2), histtype='bar')
plt.title('histtype - bar')
plt.figure()

plt.hist((weight, weight2), histtype='barstacked')
plt.title('histtype - barstacked')
plt.figure()

plt.hist((weight, weight2), histtype='step')
plt.title('histtype - step')
plt.figure()

plt.hist((weight, weight2), histtype='stepfilled')
plt.title('histtype - stepfilled')
plt.show()
```

#### Matplotlib 텍스트 삽입하기
text() 함수는 그래프의 적절한 위치에 텍스트를 삽입하도록 함   
plt.text(x좌표, y좌표, "텍스트")  
fontdict={} : 텍스트 스타일 설정  
rotation = 숫자 : 텍스트 회전(각도값)  
bbox = {} : 텍스트 상자 스타일 지정

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16}

font2 = {'family': 'Times New Roman',
        'color':  'blue',
        'weight': 'bold',
        'size': 12,
        'alpha': 0.7}

font3 = {'family': 'Arial',
        'color':  'forestgreen',
        'style': 'italic',
        'size': 14}

box1 = {'boxstyle': 'round',
        'ec': (1.0, 0.5, 0.5),
        'fc': (1.0, 0.8, 0.8)}

box2 = {'boxstyle': 'square',
        'ec': (0.5, 0.5, 1.0),
        'fc': (0.8, 0.8, 1.0),
        'linestyle': '--'}

box3 = {'boxstyle': 'square',
        'ec': (0.3, 1.0, 0.5),
        'fc': (0.8, 1.0, 0.5),
        'linestyle': '-.',
        'linewidth': 2}

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.text(-3.0, 0.15, 'np.random.randn()', fontdict=font1, rotation=85, bbox=box1)
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict=font2, rotation=-60, bbox=box2)
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict=font3, bbox=box3)
plt.show()
```

#### Matplotlib 다양한 패턴 채우기
bar(hatch = '종류') : {'/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*'} 막대 안에 넣을 패턴 지정  
hatch 문자열의 개수를 한 개 이상 입력함으로써 패턴의 밀도를 지정  

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['hatch.linewidth'] = 3

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "/" * i

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='skyblue', edgecolor='black')

    for bar in bars:
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```
